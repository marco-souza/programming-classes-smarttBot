## Python

Python é uma linguagem de programação __interpretada__ de __alto nível__, ou seja, mais proxima da __linguagem humana__ que a __linguagem de máquina__. 

Criada por __Guido Van Rossum__ em 1991, Python tem esse nome em homenagem ao grupo de comédia britânica __Monty Python Flying Circus__, o qual Guido é fã.

A linguagem foca em __legibilidade e simplicidade__ de código, __tornando a vida mais fácil de quem mantem o software funcionando__. Os principios por trás da linguagem podem ser lidos no __"The Zen Of Python"__ digitando o comando no terminal/console:

```
import this
```

## Por onde começar?

### 1. REPL (Read, Evaluate, Print, Loop)

__REPL__ é um meio interativo de utilizar o Python. É uma ótima maneira de testar e verificar o funcionamento de funções sem precisar de criar um __script__.

Para abrir o __REPL__ no Linux e Mac, basta abrir um terminal e digitar 
```
python
```

No Windows, após a instalação, você poderá usar o CMD ou o __IDLE__(recomendado), um programa que interpreta código python como linha de comando com algumas funconalidades adicionais como __auto-complete__, __sugestões__, __indicação de parâmetros__ e etc.




### 2. Script

__Script__ basicamente é um arquivo de texto. Neste arquivo de texto vem o seu programa e ele pode ser executado facilmente. Este arquivo deve ter a extensão __.py__

Para executar seu script na linha de comando, utilize:

```
python my-script.py
```

Ou aplique permissões de execução para que seu arquivo seja aberto assim:

```
./my-script.py
```

### 3. Shebang

Você pode criar um "executável" pra seu código sem a extenção .py, pPara isso utilizamos o __shebang__. 

Shebang é basicamente um cabeçalho que permite arquivos de código interpretável definam qual __programa ele irá utilizar para executar o código__ ao usar para abrir. Para definir o shebang, colocamos na primeira linha do arquivo: `#! /path/to/interpreter`, por exemplo, para rodar um código python, no cabeçalho do arquivo coloque a seguinte linha:

```
#! env python
```
